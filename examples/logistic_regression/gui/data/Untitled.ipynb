```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
df = pd.read_csv('test_2.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

df
```

```python
(np.sum(df['alarm'])/10000)
```

```python
model = smf.logit(formula="alarm~delivery_type", data=df)
results = model.fit(maxiter=40)
results.summary()
```

```python
distance = pd.DataFrame([
    { 'delivery_type': 1 , 'intercept': 1},
    { 'delivery_type': 0, 'intercept': 1},
])


distance['predicted'] = results.predict(distance)
distance
```

```python
model = smf.logit(formula="alarm~delivery_type+length", data=df)
results = model.fit(maxiter=40)
results.summary()
```

```python
content = pd.DataFrame([
    {'delivery_type': 0, 'length': 95.67952366521307},
    {'delivery_type': 0, 'length': 10},
    {'delivery_type': 1,'length': 89.86199664977082},
    {'delivery_type': 1,'length': 10},
    {'delivery_type': 1,'length': 100},
])

content['predicted'] = results.predict(content)
content
```

```python
model = smf.logit(formula="S1~delivery_location", data=df)
results = model.fit(maxiter=40)
results.summary()
```

```python
country = pd.DataFrame([
    { 'delivery_location': 'Spain' },
    { 'delivery_location': 'Brazil' },
    { 'delivery_location': 'Cuba' },
    { 'delivery_location': 'Mexico'}

])

country['predicted'] = results.predict(country)
country
```
