```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('random_complex.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

df
```

```python
drop_cols = ['country', 'distance', 'content', 'amount', 'S2', 'S3', 'alarm']

df.drop(drop_cols, axis=1)
```

```python
# What effect does the country have on whether alarm was set off?
model = smf.logit(formula='alarm ~ S1+S2+S3', data=df)
results = model.fit(maxiter=40)
```

```python
transport = pd.DataFrame([
    { 'plane_transport': 1 },
    { 'plane_transport': 0 },

])


transport['predicted'] = results.predict(transport)
```
